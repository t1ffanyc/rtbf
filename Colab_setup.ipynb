The training logs for the above models are stored in the same folder, as training_log.txt in each folder. We will now generate a graph to visualize the loss across training. The first 5 epochs were excluded from this graph to improve visibility of differences between the models.

```python
from IPython.display import Image

!python scripts/plot_training_loss_comparison.py \
  logs/2026-03-11/17-29-35/training_log.txt logs/2026-03-12/05-01-28/training_log.txt logs/2026-03-12/20-48-08/training_log.txt logs/2026-03-12/08-33-56/training_log.txt \
  --labels "baseline CNN, pure RNN, CNN + RNN hybrid, CNN + transformer hybrid" \
  -o comparison.png

display(Image("comparison.png"))
```

We believe the transformer model is overfitting. We will plot the val CER and the training loss on the same graph. We can observe that though the val CER stops improving, the training loss keeps decreasing, which suggests overfitting.
